# Visuals

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for classifiers' average F1 and accuracy scores
classifiers = ['Dummy Classifier', 'Paper Classifier', 'Top Text Model', 'Custom Pipeline']
average_f1_scores = [0.17938, 0.53995, 0.691044, 0.719147]  #  F1 scores
average_accuracy_scores = [0.1761, 0.556889, 0.699298, 0.72701]  # accuracy scores

f1_std_devs = [0.0317, 0.01117859, 0.009947, 0.02038]  #  F1 score standard deviations
accuracy_std_devs = [0.0203, 0.012957, 0.009880, 0.0195]  # accuracy score standard deviations

# Bar width for grouped bars
bar_width = 0.35
x = np.arange(len(classifiers))  # Position of bars on x-axis

# Create the plot
fig, ax = plt.subplots(figsize=(8, 6))

# Add F1 and accuracy bars
bars1 = ax.bar(
    x - bar_width / 2,
    average_f1_scores,
    bar_width,
    yerr=f1_std_devs,
    capsize=5,
    label='F1 Score',
    color='skyblue',
    alpha=0.8
)
bars2 = ax.bar(
    x + bar_width / 2,
    average_accuracy_scores,
    bar_width,
    yerr=accuracy_std_devs,
    capsize=5,
    label='Accuracy',
    color='salmon',
    alpha=0.8
)

# Add labels and title
ax.set_xlabel('Classifiers', fontsize=12)
ax.set_ylabel('Scores', fontsize=12)
ax.set_title('Average Cross-Validation F1 and Accuracy Scores', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(classifiers)
ax.legend()

# Annotate bar values
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}', 
                    xy=(bar.get_x() + bar.get_width() / 2, height+0.02),
                    xytext=(0, 3),  # Offset for the text
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=10)

# Show plot
plt.tight_layout()
plt.show()
```
